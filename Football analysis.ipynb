```python
import statistics 
from statistics import StatisticsError
import csv
import math
```

```python
def calc_points(key,number):
    if(key=='passyds_values'):
        return math.floor(number/25)
    if(key=='pass.tds_values'):
        return number*6
    if(key=='pass.ints_values'):
        return -2*number
    if(key=='pass.twoptm_values'):
        return 2*number
    if(key=='rushyds_values'):
        return math.floor(number/10)
    if(key=='rushtds_values'):
        return number*6
    if(key=='rush.twoptm_values'):
        return 2*number
    if(key=='recyds_values'):
        return math.floor(number/10)
    if(key=='rec.tds_values'):
        return number*6
    if(key=='rec.twoptm_values'):
        return number*2
    if(key=='fgm_values'):
        return 3*number
    if(key=='xpmade_values'):
        return number
    return 0
```

```python
player_long = {}
column_headers = []

# input the csv file
with open('player.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    first_row = True
    #for each row in csv
    for row in csv_reader:
        
        #get the column headers from the first row
        if(first_row):
            column_headers = row
            first_row=False
            continue
            
        #if new player
        if(row[6] not in player_long):
            player_information = {}
            for column_number in range(len(column_headers)):
                if(column_number>6):
                    player_information[column_headers[column_number]]=row[column_number]
                    player_information[str(column_headers[column_number])+"_values"]=[float(row[column_number])]
            
            player_information['games_played']=1
            
            player_long[row[6]]=player_information
            
        #if player already in dictionary
        else:
            player_information = player_long[row[6]]
            for column_number in range(len(column_headers)):
                if(column_number<=6):
                    continue
                if(column_headers[column_number]=='games'):
                    continue
                player_information[column_headers[column_number]]=float(row[column_number])+float(player_information[column_headers[column_number]])
                player_information[str(column_headers[column_number])+"_values"].append(float(row[column_number]))
            player_information['games_played']=int(player_information['games_played'])+1
            player_long[row[6]]=player_information
```

```python
player_points={}
for key in player_long:
    count=0
    last_information = ""
    
    for information in player_long[key]:
        if(key not in player_points):
                player_points[key]={}
                player_points[key]['ppg']=[]
                for keys in range(len(player_long[key]['pass.att_values'])):
                    player_points[key]['ppg'].append(0)
        if(count%2==1):
            for number in range(len(player_long[key][information])):
                 player_points[key]['ppg'][number]= player_points[key]['ppg'][number]+calc_points(information,player_long[key][information][number])
        
        last_information = information
        count = count + 1
    player_points[key]['games']=player_long[key]['games_played']

    
```

```python
for key in player_points:
    try:
        player_points[key]['ppg.mean']=statistics.mean(player_points[key]['ppg'])
        player_points[key]['ppg.std']=statistics.stdev(player_points[key]['ppg'])
        player_points[key]['ppg.median']=statistics.median(player_points[key]['ppg'])
        player_points[key]['ppg.min']=min(float(ppg) for ppg in player_points[key]['ppg'])
        player_points[key]['ppg.max']=max(float(ppg) for ppg in player_points[key]['ppg'])
    except:
        player_points[key]['ppg.mean'] = player_points[key]['ppg'][0]
        player_points[key]['ppg.std']=0
        player_points[key]['ppg.min']=player_points[key]['ppg'][0]
        player_points[key]['ppg.median']=player_points[key]['ppg'][0]
        player_points[key]['ppg.max']=player_points[key]['ppg'][0]
        
```

```python
player_rankings={}
# input the csv file
with open('rankings.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    first_row = True
    #for each row in csv
    for row in csv_reader:
        
        if(first_row):
            first_row=False
            continue
        
        player_rankings[str(row[0])+str(row[1])]={
            'cbs.rank':int(row[2]),
            'cbs.cap':int(row[3])
        }
```

```python
player_stats={}
for players in player_points:
    #if(float(player_points[players]['games'])<10  or float(player_points[players]['ppg.mean'])<15):
    #    continue
    player_stats[players]={}
    player_stats[players]['ppg.mean']=player_points[players]['ppg.mean']
    player_stats[players]['ppg.std']=player_points[players]['ppg.std']
    player_stats[players]['ppg.min']=player_points[players]['ppg.min']
    player_stats[players]['ppg.max']=player_points[players]['ppg.max']
    player_stats[players]['ppg.median']=player_points[players]['ppg.median']
    if(players in player_rankings):
        player_stats[players]['cbs.rank']=player_rankings[players]['cbs.rank']
        player_stats[players]['ppg.mean.adj']=player_points[players]['ppg.mean']+(player_rankings[players]['cbs.cap']/10)-float(player_points[players]['ppg.std']/10)
    else:
        player_stats[players]['cbs.rank']=201
        player_stats[players]['ppg.mean.adj']=float(player_points[players]['ppg.mean'])+3
```
